```python
import numpy as np
```

```python
x = []
y = []
with open("deep learning.csv") as f: 
    for line in f: 
        t = line.replace("\n","").split(",")
        xx = []
        yy = []
        for i in range(0, 11):
            try:
                x1 = []
                for xt in t[1:8]:
                    x1.append(float(xt)/200.0)
                x1.append(1) if float(t[8+i])> 5 else x1.append(0)
                x.append(x1)
                y.append(1) if float(t[9+i])> 5 else y.append(0)
            except ValueError:
                pass
```

```python
y
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv1D,MaxPooling1D
```

```python
model = Sequential()
```

```python
model.add(Dense(units=40, input_dim = 8,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=30, kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=1, kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
train_history=model.fit(x=np.array(x[:]),y=np.array(y[:]),validation_split=0.2,epochs=20,batch_size=40,verbose=2)
```

```python
prediction=model.predict_classes(np.array([[0.0, 0.28, 0.005, 0.125, 0.005, 0.8009999999999999, 0.245,0]]))
```
